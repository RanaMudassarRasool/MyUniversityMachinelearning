# Machine learning


Machine learning is a type of artificial intelligence (AI) that allows software applications to become more accurate in predicting outcomes without being explicitly programmed to do so. Machine learning algorithms use historical data as input to predict new output values.



# There are four main types of machine learning:

**Supervised learning**: In supervised learning, the algorithm is trained on a set of data that has been labeled with the desired output. For example, a supervised learning algorithm could be trained on a dataset of images that have been labeled as either cats or dogs. The algorithm would then learn to identify new images as cats or dogs based on the patterns it has learned from the training data.


**Unsupervised learning:** In unsupervised learning, the algorithm is trained on a set of data that has not been labeled. The algorithm then tries to find patterns and relationships in the data on its own. For example, an unsupervised learning algorithm could be used to segment customers into different groups based on their purchase history.

**Semi-supervised learning:** Semi-supervised learning is a hybrid of supervised and unsupervised learning. The algorithm is trained on a set of data that is partially labeled. The algorithm then uses the labeled data to learn patterns, which it then uses to classify the unlabeled data.


**Reinforcement learning:** In reinforcement learning, the algorithm learns by trial and error. The algorithm is placed in an environment and is given a goal. The algorithm then takes actions and receives rewards or punishments based on the outcome of its actions. The algorithm then learns to take actions that lead to rewards and avoid actions that lead to punishments.

# Module 2: Supervised Learning

Supervised learning is a type of machine learning in which the algorithm is trained on a set of data that has been labeled with the desired output. For example, a supervised learning algorithm could be trained on a dataset of images that have been labeled as either cats or dogs. The algorithm would then learn to identify new images as cats or dogs based on the patterns it has learned from the training data

**Common supervised learning algorithms:**

**Linear regression:** Linear regression is a supervised learning algorithm that is used to predict continuous values, such as the price of a house or the number of customers who will visit a store on a given day.

**Logistic regression:** Logistic regression is a supervised learning algorithm that is used to predict binary values, such as whether or not a customer will click on an ad or whether or not a patient has a disease.

**Decision trees:** Decision trees are supervised learning algorithms that are used to classify data into different categories.

**Support vector machines:** Support vector machines are supervised learning algorithms that are used to classify data into different categories and to detect anomalies in data.

**Ensemble methods:** Ensemble methods are supervised learning algorithms that combine the predictions of multiple algorithms to produce a more accurate prediction.

**Supervised learning workflow:**



**Collect data:** The first step is to collect a dataset of labeled data. The data should be representative of the data that the algorithm will be used to predict on in the future.

**Prepare the data:** The next step is to prepare the data for training. This may involve cleaning the data, removing outliers, and scaling the data.

**Choose a model:** The next step is to choose a supervised learning algorithm. The choice of algorithm will depend on the type of data and the desired output.

**Train the model:** The next step is to train the model on the prepared data. This involves feeding the data to the algorithm and allowing it to learn the patterns in the data.

**Evaluate the model:** Once the model is trained, it is important to evaluate its performance on a held-out test set. This will give you an idea of how well the model will generalize to unseen data.

**Deploy the model:** Once the model is evaluated and you are satisfied with its performance, you can deploy the model to production. This means making the model available to users so that they can make predictions using the model.

# When choosing a supervised learning algorithm, there are a few factors to consider:

**Type of data:** Is the data categorical or continuous? If the data is categorical, then you will need to choose a classification algorithm. If the data is continuous, then you will need to choose a regression algorithm.

**Desired output:** Do you want to predict a continuous value (e.g., the price of a house) or a categorical value (e.g., whether or not a customer will click on an ad)? The type of output will also determine the type of algorithm that you need to choose.

**Complexity of the data:** Is the data linear or non-linear? If the data is linear, then you can choose a simpler algorithm. If the data is non-linear, then you will need to choose a more complex algorithm.

# Linear regression

Linear regression is a supervised learning algorithm that is used to predict continuous values. It is one of the simplest and most widely used machine learning algorithms.

Linear regression works by fitting a linear equation to the data. The equation is of the form:

                                         # y = mx + b

where:

- y is the predicted value
- x is the independent variable
- m is the slope of the line
- b is the y-intercept

The slope and y-intercept of the line are estimated by minimizing the sum of squared errors (SSE) between the predicted values and the actual values.

Linear regression can be used to predict a wide variety of continuous values, such as:

- House prices
- Customer churn
- Medical costs
- Product sales
- Weather forecasts

The dataset used in linear regression should be a dataset of paired values, where one value is the dependent variable (y) and the other value is the independent variable (x). The dataset should also be representative of the data that the algorithm will be used to predict on in the future.

Here is an example of a dataset that could be used in linear regression:


x | y
-- | --
10 | 20
20 | 30
30 | 40
40 | 50

This dataset could be used to predict the value of y for a given value of x. For example, if we wanted to predict the value of y for x = 50, we would use the linear regression equation to predict that y = 60.

Linear regression is a powerful tool that can be used to solve a wide variety of problems. It is a good choice for problems where the relationship between the independent and dependent variables is linear.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load the house price dataset
df = pd.read_csv('USA_Housing.csv')
```

```python
df.head(3)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
               'Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

# Train Test Split

```python



# Split the dataset into training and test datasets
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)
```

# Creating and Training the Model

```python

# Create a linear regression model
model = LinearRegression()

# Train the model on the training dataset
model.fit(X_train, y_train)

```

```python

# Make predictions on the test dataset
y_pred = model.predict(X_test)

# Evaluate the model performance on the test dataset
mse = np.mean((y_pred - y_test)**2)
print('Mean squared error:', mse)

```
