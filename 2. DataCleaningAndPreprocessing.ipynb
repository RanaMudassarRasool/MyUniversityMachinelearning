2. **dropna():** This function removes rows or columns containing missing values based on specified criteria.

**dropna():** This function removes rows or columns containing missing values based on specified criteria.
1. axis='index' removes rows with missing values.

2. axis='columns' removes columns with missing values.

3. how='any' removes a row or column if any value is missing.

4. how='all' removes a row or column only if all values are missing.

5. thresh=n removes a row or column if it has less than n non-missing values.

6. subset=[list of columns] restricts the check for missing values to specific columns.

**axis='index': Drops rows that contain missing values. This is the default behavior.**

```python

df = pd.read_csv('employees.csv')

print("======= before  =========")
print(df.shape) 
print(df.isnull().sum()) 

# Drop rows with any missing values
df.dropna(axis='index', inplace=True)
print("======= After  =========")
# Check if there are still missing values 
print(df.isnull().sum())
```

**axis='columns': Drops columns that contain missing values.**

```python
import pandas as pd

df = pd.read_csv('employees.csv')
# Print original DataFrame shape

print("======= before  =========")
print(df.shape) 
print(df.isnull().sum())

print("======= After  =========")
# Drop columns with any missing values
df.dropna(axis='columns', inplace=True)
# Print new DataFrame shape  
print(df.shape)

print(df.isnull().sum())
```

**how='any': If any value in the row/column is missing, drop the entire row/column. This is the default**.

```python
import pandas as pd

df = pd.read_csv('employees.csv')

print("======= BEFORE ========")
print(df.shape)
print(df.isnull().sum()) 

# Drop columns with any missing values
df.dropna(axis='columns', how='any', inplace=True)

print("\n======= AFTER ========")
print(df.shape) 
print(df.isnull().sum())
```

```python
import pandas as pd

df = pd.read_csv('employees.csv')

print("======= BEFORE =======")
print(df.shape)
print(df.isnull().sum())

# Only drop rows with missing values in columns
df.dropna(axis='index', subset=['First Name','Gender'], inplace=True)

print("\n======= AFTER =======")
print(df.shape)
print(df.isnull().sum())
```

**Interpolation**: Fill missing values by interpolating between neighboring points. Pandas provides interpolation methods like interpolate() and spline() for this.

```python
import pandas as pd

df = pd.read_csv('employees.csv')

print("======= BEFORE =======")
print(df.shape)
print(df.isnull().sum())

# Interpolate missing values  
df = df['Salary'].interpolate()

print("\n======= AFTER =======")
print(df.shape)  
print(df.isnull().sum())
df.info()
```

# Identifying and Handling Outliers:

Outliers are extreme values that deviate significantly from the rest of the data. They can skew the results of machine learning models. Pandas offers methods to detect and address outliers:

**describe()**: This function provides summary statistics, including quartiles, which can help identify potential outliers.

**boxplot()**: Boxplots visually represent the distribution of data and highlight outliers.

**zscore()**: This function calculates the z-score, a measure of how far a value is from the mean. Large z-scores may indicate outliers.

Outliers can be handled by removing them, imputing them with less extreme values, or using robust models less sensitive to outliers.

```python
pip install scipy
```

```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv('employees.csv')

print(df.describe()) 

# Create boxplots 
df.boxplot(column=['Salary', 'Bonus %'])

# Calculate z-scores
z = np.abs(stats.zscore(df))

# Filter outlier rows by z-score threshold
df = df[(z < 3).all(axis=1)]  

print(df.describe())
```

# How to Handle Missing Data in Python?

Handling missing data in Python can be crucial for effective data analysis. Here's a simplified explanation of the process in five easy steps:

**Step 1: Why Fill in the Missing Data?**
Missing data can lead to inaccurate analysis and predictions. It's essential to handle missing values to ensure the integrity of your data and the reliability of your results.

**Step 2: How to Know If the Data Has Missing Values?**
You can use Python libraries like Pandas to check for missing values. 
The **isna() or isnull()** function helps identify where data is missing.

**Step 3: Different Methods of Dealing With Missing Data**
There are several methods to handle missing data:
 1. **Deleting the Column with Missing Data**: If a column has too many missing values and is not crucial, you can remove it using the drop() function in Pandas.

 2. **Deleting the Row with Missing Data:** If only a few rows have missing values, you can delete them using the dropna() function.

 3. **Filling the Missing Values - Imputation:** You can fill missing values with reasonable estimates. For example, you can replace missing values with the mean, median, or mode of the column using the fillna() function.

 4. **Other Imputation Methods:** You can also use more advanced imputation techniques like forward-fill, backward-fill, or interpolation methods depending on your data and domain.

 5. **Filling with a Regression Model:** If you have a substantial amount of data, you can use regression models to predict and fill missing values.

**Step 4: Conclusion**
Choosing the right method depends on your data and the problem you're trying to solve. Be mindful of the potential impact on your analysis and results.

**Step 5: Frequently Asked Questions**

 **Why is missing data a problem?** Missing data can lead to biased or incorrect conclusions in data analysis.

 **When should you delete missing data?** Consider deleting when it won't significantly affect your analysis and there are few missing values.

 **When should you impute missing data?** Use imputation when you want to retain as much data as possible or deleting would result in significant information loss.

 **What imputation method should I use?** It depends on your data and the nature of the missing values. Common methods include mean, median, mode, or more advanced techniques like regression-based imputation.

 **How can I automate missing data handling?** You can create functions or use Python libraries like Scikit-learn to automate missing data handling in larger datasets.
 
In summary, dealing with missing data in Python involves understanding the nature of your data, using appropriate methods for imputation or deletion, and making informed choices to ensure the quality of your analysis.
